```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline


def printAllNA(dataset):
    print(dataset[dataset.isna().any(axis=1)])

def minMaxScaling(dataset_Column):
    
    minData = dataset_Column.min()
    maxData = dataset_Column.max()
    
    return dataset_Column.apply(lambda x : (x-minData)/(maxData-minData))

def zScores(dataset):
    miu = dataset.mean()
    standard_dev = dataset.std()

    dataset = dataset.select_dtypes(np.number)
    
    for column in dataset.columns:
        dataset[column] = (dataset[column] - dataset[column].mean())/ dataset[column].std()

    
    return dataset



def removeOutliers(dataset):
    z_scores = zScores(dataset)
    return dataset[(np.abs(z_scores) <= 3).all(axis=1)]


def convertProximityToColor(ocean_proximity):
        dictColor = {
            'NEAR BAY':[1.0,0,0],
            '<1H OCEAN':[0,1.0,0],
            'INLAND' : [0,0,1.0],
            'NEAR OCEAN' : [0.5,0.5,0],
            'ISLAND' : [0,0.5,0.2]
        }
        
    
        return ocean_proximity.apply(lambda x : dictColor[x])
        


    
dataset_Path ="Datasets/California Housing/housing.csv"

housing_Dataset = pd.read_csv(dataset_Path)

housing_Dataset = housing_Dataset.dropna()
housing_Dataset_wo_Outliers = removeOutliers(housing_Dataset)

#housing_Dataset_wo_Outliers = housing_Dataset_wo_Outliers[housing_Dataset_wo_Outliers["median_house_value"] <= 500000.1 ]
print(housing_Dataset_wo_Outliers.head())
print(housing_Dataset_wo_Outliers.describe())
```
